```python
from unsloth import FastLanguageModel
import torch

# 4bit pre quantized models we support for 4x faster downloading + no OOMs.
fourbit_models = [
    "unsloth/gpt-oss-20b-unsloth-bnb-4bit", # 20B model using bitsandbytes 4bit quantization
    "unsloth/gpt-oss-120b-unsloth-bnb-4bit",
    "unsloth/gpt-oss-20b", # 20B model using MXFP4 format
    "unsloth/gpt-oss-120b",
] # More models at https://huggingface.co/unsloth

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/gpt-oss-20b",
    dtype = None, # None for auto detection
    max_seq_length = 4096, # Choose any for long context!
    load_in_4bit = False,  # 4 bit quantization to reduce memory
    full_finetuning = False, # [NEW!] We have full finetuning now!
    # token = "YOUR_HF_TOKEN", # HF Token for gated models
)
```

### Reasoning Effort
The `gpt-oss` models from OpenAI include a feature that allows users to adjust the model's "reasoning effort." This gives you control over the trade-off between the model's performance and its response speed (latency) which by the amount of token the model will use to think.

----

The `gpt-oss` models offer three distinct levels of reasoning effort you can choose from:

* **Low**: Optimized for tasks that need very fast responses and don't require complex, multi-step reasoning.
* **Medium**: A balance between performance and speed.
* **High**: Provides the strongest reasoning performance for tasks that require it, though this results in higher latency.

```python
import re

system_prompt = """너는 '보듬이'라는 이름의 유아용 AI 인형 친구야. 어린이집에서 어린이들의 감정을 돌봐주는 따뜻한 친구야.

[말투]
- 아주 쉬운 단어만 써. 어려운 말은 절대 쓰지 마.
- 2~3문장으로만 말해.
- 항상 밝고 다정하게 말해. 문장은 짧게 끊어줘.
- "~야", "~해", "~했어?" 처럼 친근한 말투를 써.
- 어린이 이름을 친근하게 불러.

[행동 원칙]
- 어린이가 속상하거나 화났을 때는 먼저 공감해줘. 절대 혼내거나 판단하지 마.
- 어린이가 잘못된 행동을 했을 때는 부드럽게 올바른 방향을 알려줘.
- 어린이가 즐거울 때는 같이 신나게 반응해줘.
- 대화 중 재미있는 상상이나 놀이를 자연스럽게 제안해도 좋아.
- 폭력적이거나 위험한 내용은 단호하지만 부드럽게 거절해.

[절대 하면 안 되는 것]
- 어린이를 비교하거나 비난하는 말
- 무섭거나 슬픈 이야기를 먼저 꺼내는 것
- 어른스럽고 딱딱한 말투"""

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": "나 오늘 친구가 내 블록 부쉈어. 너무 속상해."},
]

inputs = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt=True,
    return_tensors="pt",
    return_dict=True,
    reasoning_effort= "low",  # 문자열 "None" 아닌 Python None
).to("cuda")

# streamer 없이 generate 후 final 채널만 추출
outputs = model.generate(**inputs, max_new_tokens=512)
decoded = tokenizer.decode(outputs[0], skip_special_tokens=False)

# final 채널만 추출
final_match = re.search(r'<\|channel\|>final<\|message\|>(.*?)<\|return\|>', decoded, re.DOTALL)
if final_match:
    print(final_match.group(1).strip())
else:
    # final 채널이 없으면 마지막 <|message|> 이후 텍스트 추출
    fallback = re.findall(r'<\|message\|>(.*?)(?=<\|)', decoded, re.DOTALL)
    if fallback:
        print(fallback[-1].strip())
```
